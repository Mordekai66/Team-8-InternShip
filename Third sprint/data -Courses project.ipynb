### Model without interface

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import pandas as pd


data = pd.read_excel(r"C:\Users\Workstation Shop\Desktop\data - project.xlsx")


le_a = LabelEncoder()
le_b = LabelEncoder()
le_c = LabelEncoder()
le_d = LabelEncoder()

data["A"] = le_a.fit_transform(data['A'])
data["B"] = le_b.fit_transform(data['B'])
data["C"] = le_c.fit_transform(data['C'])
data["D"] = le_d.fit_transform(data['D'])


x_train = data[["A","B","C","D"]]

kmeans = KMeans(n_clusters=48, random_state=42)
kmeans.fit(x_train)

new_data = {'a': 'programming', 'b': 'python', 'c': 'Beginner', 'd': 'free'}

new_a_encoded = le_a.transform([new_data['a']])[0]
new_b_encoded = le_b.transform([new_data['b']])[0]
new_c_encoded = le_c.transform([new_data['c']])[0]
new_d_encoded = le_d.transform([new_data['d']])[0]

test= pd.DataFrame([[new_a_encoded, new_b_encoded, new_c_encoded, new_d_encoded]],
                     columns=['A', 'B', 'C', 'D'])

pridict_cluster = kmeans.predict(test)[0]

print("Available courses:")

for i, cluster in enumerate(kmeans.labels_):
    if cluster == pridict_cluster:
        print(f"Course: {data['E'][i]}")
```

### interface code with model

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
import joblib
import gradio as gr


data = pd.read_excel(r"C:\Users\Workstation Shop\Desktop\data - project.xlsx")


category_list = data["A"].unique().tolist()
programming_language_list = data["B"].unique().tolist()
difficulty_list = data["C"].unique().tolist()
free_paid_list = data["D"].unique().tolist()


le_a = LabelEncoder()
le_b = LabelEncoder()
le_c = LabelEncoder()
le_d = LabelEncoder()

data["A"] = le_a.fit_transform(data['A'])
data["B"] = le_b.fit_transform(data['B'])
data["C"] = le_c.fit_transform(data['C'])
data["D"] = le_d.fit_transform(data['D'])


x_train = data[["A","B","C","D"]]


model = KMeans(n_clusters=48, random_state=42)
model.fit(x_train)

joblib.dump(model, "kmeans_model.pt")
joblib.dump(le_a, "le_a.pt")
joblib.dump(le_b, "le_b.pt")
joblib.dump(le_c, "le_c.pt")
joblib.dump(le_d, "le_d.pt")


model = joblib.load("kmeans_model.pt")
le_a = joblib.load("le_a.pt")
le_b = joblib.load("le_b.pt")
le_c = joblib.load("le_c.pt")
le_d = joblib.load("le_d.pt")



def predict(a, b, c, d):
    if a != "programming":
        b = "NAN"
    new_a_encoded = le_a.transform([a])[0]
    new_b_encoded = le_b.transform([b])[0]
    new_c_encoded = le_c.transform([c])[0]
    new_d_encoded = le_d.transform([d])[0]

    test= pd.DataFrame([[new_a_encoded, new_b_encoded, new_c_encoded, new_d_encoded]],
                     columns=['A', 'B', 'C', 'D'])

    pridict_cluster = model.predict(test)[0]

    result = "Courses based on your options:\n"
    for i, cluster in enumerate(model.labels_):
        if cluster == pridict_cluster:
            result += f"Course: {data['E'][i]}\n"
    return result

pro = gr.Interface(
    fn=predict,
    inputs=[
        gr.Dropdown(label="Category", choices=category_list),
        gr.Dropdown(label="Programming Language", choices=programming_language_list),
        gr.Dropdown(label="Difficulty", choices=difficulty_list),
        gr.Dropdown(label="Free/Paid", choices=free_paid_list)],
    outputs="text"
)
pro.launch()
```
